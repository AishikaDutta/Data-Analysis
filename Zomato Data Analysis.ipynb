```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe.head())
```

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)
 
dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
sns.countplot(x=dataframe['listed_in(type)'], hue=dataframe['listed_in(type)'], palette='viridis')
plt.xlabel("Type of Restaurant")
plt.ylabel("Count")
plt.title("Restaurant Types with Category")
plt.show()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="black", marker="o",markerfacecolor="red")
plt.xlabel("Type of restaurant", c="black", size=20)
plt.ylabel("Votes", c="black", size=20)
plt.title("Votes by Type of Restaurant", color="black", size=22)
plt.grid(True)
```

```python
max_votes = dataframe['votes'].max()
restaurant_with_max_votes = dataframe.loc[dataframe['votes'] == max_votes, 'name']
 
print("Restaurant(s) with the maximum votes:")
print(restaurant_with_max_votes)
```

```python
sns.countplot(x=dataframe['online_order'], hue=dataframe['online_order'], palette='viridis')
plt.xlabel("Online Order")
plt.ylabel("Count")
plt.title("Online Orders")
plt.show()
```

```python
plt.hist(dataframe['rate'],bins=5,color="lightgreen",edgecolor="black")
plt.title("Ratings Distribution")
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(x=dataframe['approx_cost(for two people)'], hue=dataframe['approx_cost(for two people)'], palette='viridis')
plt.xlabel("Costing for 2 people")
plt.ylabel("Count")
plt.title("Approx_Cost")
plt.show()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```
